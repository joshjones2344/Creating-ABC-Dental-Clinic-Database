```python
import pandas as pd
from datetime import datetime
import time


# Load JSON data
with open('Appointment.json', 'r') as json_file:
    data = json.load(json_file)

# Convert JSON data to a DataFrame
df = pd.DataFrame(data)

# Convert Appn_Date to datetime
df['Appn_Date'] = pd.to_datetime(df['Appn_Date'])
# convert time_start and time_end with am and pm to 24 hour format
df['Time_start'] = pd.to_datetime(df['Time_start'], format='%I:%M%p').dt.time
df['Time_end'] = pd.to_datetime(df['Time_end'], format='%I:%M%p').dt.time
```

```python
df
#save as csv to current directory
df.to_csv('Appointements.csv', index=False)
```

```python
import pandas as pd
import json

# Load JSON data
with open('Dentist.json', 'r') as json_file:
    data = json.load(json_file)

# Convert JSON data to a DataFrame
df = pd.DataFrame(data)

# Save DataFrame as CSV
df.to_csv('Dentists.csv', index=False)
```

```python
#convert DentistID to int
df['DentistID'] = df['DentistID'].astype(int)

#convert Phone to int with 9 digits
df['Phone'] = df['Phone'].astype(str).str[:9].astype(int)

df.dtypes
df.to_csv('Dentists.csv', index=False)
```
